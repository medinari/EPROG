### Fehlerbehandlung

Die folgenden Fehlerfälle müssen im Spiel berücksichtigt und entsprechend behandelt werden:

---

#### 1. **Ungültige Eingabe des Spielers**
Falls ein Spieler eine Eingabe tätigt, die keine gültige Spaltennummer (1–7) ist:
- Zeigen Sie eine Fehlermeldung an: `"Ungültige Eingabe. Bitte wählen Sie eine Spalte zwischen 1 und 7."`
- Fordern Sie den Spieler auf, die Eingabe zu wiederholen.

**Beispiel:**

```plaintext

Spieler 1 (X), bitte wählen Sie eine Spalte (1-7): 9
Ungültige Eingabe. Bitte wählen Sie eine Spalte zwischen 1 und 7.
Spieler 1 (X), bitte wählen Sie eine Spalte (1-7): 4



#### 2. **Spalte ist bereits voll**
Falls ein Spieler eine Spalte wählt, die bereits voll ist:
- Zeigen Sie eine Fehlermeldung an: `"Die gewählte Spalte ist voll. Bitte wählen Sie eine andere Spalte."`
- Fordern Sie den Spieler auf, die Eingabe zu wiederholen.

**Beispiel:**

```plaintext

Spieler 2 (O), bitte wählen Sie eine Spalte (1-7): 4
Die gewählte Spalte ist voll. Bitte wählen Sie eine andere Spalte.
Spieler 2 (O), bitte wählen Sie eine Spalte (1-7): 5



#### 3. **Ungültiger Datentyp**
Falls der Spieler eine Eingabe macht, die kein Integer (z. B. ein Buchstabe oder Symbol) ist:
- Zeigen Sie eine Fehlermeldung an: `"Ungültige Eingabe. Bitte geben Sie eine Zahl zwischen 1 und 7 ein."`
- Fordern Sie den Spieler auf, die Eingabe zu wiederholen.

**Beispiel:**

```plaintext

Spieler 1 (X), bitte wählen Sie eine Spalte (1-7): a

Spieler 1 (X), bitte wählen Sie eine Spalte (1-7): 3
