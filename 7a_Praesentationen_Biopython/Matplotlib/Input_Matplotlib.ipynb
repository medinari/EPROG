# Matplotlib
### Syntax basics

| **Feature**            | **Methods**               | **Description**                                                                 | **Example Code**                         |
|-------------------------|---------------------------|---------------------------------------------------------------------------------|------------------------------------------|
| **Line Color**          | `color` or `c`           | Sets the color of the plot line.                                                | `plt.plot(x, y, color='blue')`           |
| **Line Style**          | `linestyle` or `ls`      | Defines the style of the plot line (solid, dashed, etc.).                       | `plt.plot(x, y, linestyle='--')`         |
| **Line Width**          | `linewidth` or `lw`      | Sets the width of the plot line.                                                | `plt.plot(x, y, linewidth=2)`            |
| **Marker Shape**        | `marker`                | Changes the shape of the points on the plot (circle, square).                   | `plt.plot(x, y, marker='o')`             |
| **Marker Size**         | `markersize` or `ms`     | Sets the size of the markers.                                                   | `plt.plot(x, y, markersize=5)`           |
| **Marker Color**        | `markerfacecolor` or `mfc` | Sets the fill color of the markers.                                             | `plt.plot(x, y, markerfacecolor='red')`  |
| **Title**               | `plt.title()`            | Adds a title to the plot.                                                       | `plt.title('My Plot')`                   |
| **Axis Labels**         | `plt.xlabel()` / `plt.ylabel()` | Adds labels to the x and y axes.                                              | `plt.xlabel('X-axis')`                  |
| **Legend**              | `plt.legend()`           | Adds a legend to the plot.                                                      | `plt.legend(['Label'])`                  |
| **Grid**                | `plt.grid()`             | Toggles the visibility of the grid.                                             | `plt.grid(True)`                         |
| **Annotations**         | `plt.annotate()`         | Adds annotations with optional arrows pointing to data points.                  | `plt.annotate('Max', xy=(2, 1))`         |
| **Figure Size**         | `plt.figure(figsize=(w, h))` | Sets the width and height of the plot.                                       | `plt.figure(figsize=(10, 6))`            |
| **Axis Limits**         | `plt.xlim()` / `plt.ylim()` | Sets the limits for the x and y axes.                                         | `plt.xlim(0, 10)`                        |
| **Axis Ticks**          | `plt.xticks()` / `plt.yticks()` | Sets the ticks for the x and y axes.                                         | `plt.xticks([0, 5, 10])`                 |
| **Save Plot**           | `plt.savefig()`          | Saves the current figure to a file.                                             | `plt.savefig('plot.png')`                |


### Example: Basic Line Plot



```python
# Basic Line Plot

import matplotlib.pyplot as plt
import numpy as np

# Simple data 
x = [0, 1, 2, 3, 4, 5]
y = [0,2, 4, 6, 8, 10]

# Create a line plot
plt.plot(x, y, label="2x", color = "red", marker = '^', markersize = 5, linestyle = '--') 

# Shorthand notation
#plt.plot(x, y,'r.--', label ='2x')

# Title and labels
plt.title("Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# Setting up ticks on the x and y axis
#x_ticks = np.arange(0, 10)
#y_ticks = np.arange(0, 20, 2)

#plt.xticks(x_ticks)
#plt.yticks(y_ticks)

# Making a second line using numpy
#x2 = np.arange(0, 4.5, 0.5)
#plt.plot(x2, x2**2)

plt.legend() 
plt.show() # visualizes the plot


```

```python
import matplotlib.pyplot as plt

# Data for bar chart: Favorite Snacks
categories = ['Chips', 'Chocolate', 'Popcorn', 'Ice Cream', 'Pizza']
values = [15, 20, 10, 18, 22]  # Popularity scores

# Create the bar chart
plt.bar(categories, values, color='coral', edgecolor='black')

# Add title and labels
plt.title("Snack Popularity Contest", fontsize=16)
plt.xlabel("Snacks", fontsize=12)
plt.ylabel("Popularity Score", fontsize=12)

# Annotations for the bars
for i, value in enumerate(values):
    plt.text(i, value + 0.5, f"{value} votes", ha='center', fontsize=10)

# Syntax: plt.text(x, y, text, ha, fontsize)

plt.grid(axis='y', linestyle='--', alpha=0.6)

plt.show()
```

```python
import numpy as np

# Generate random data
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

# Scatter plot
plt.scatter(x, y, c=colors, cmap='viridis', alpha=0.7)
plt.title("Scatter Plot Example")
plt.xlabel("Random X")
plt.ylabel("Random Y")
plt.colorbar(label="Color Intensity")
plt.show()
```

```python
# Data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]  # Quadratic
y2 = [2, 4, 6, 8, 10]   # Linear

# Create a figure with two Axes (subplots)
fig, ax = plt.subplots(1, 2, figsize=(10, 5))  # 1 row, 2 columns

# Plot on the first Axes
ax[0].plot(x, y1, label="Quadratic", color="blue")
ax[0].set_title("Quadratic")
ax[0].legend()

# Plot on the second Axes
ax[1].plot(x, y2, label="Linear", color="green")
ax[1].set_title("Linear")
ax[1].legend()

# Adjust layout (optional)
plt.tight_layout()

# Show the plot
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# 1. Line Plot
x = np.linspace(0, 10, 100)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.legend()
plt.savefig('line_plot_2.png')
plt.close()

# 2. Scatter Plot
x, y = np.random.rand(50), np.random.rand(50)
plt.scatter(x, y, c=x + y, cmap='viridis', alpha=0.7)
plt.colorbar(label='Intensity')
plt.savefig('scatter_plot_2.png')
plt.close()

# 3. Bar Plot
categories = ['A', 'B', 'C']
values = [20, 35, 30]
plt.bar(categories, values, color=['red', 'blue', 'green'])
plt.savefig('bar_plot_2.png')
plt.close()

# 4. Pie Chart
sizes = [25, 35, 20, 20]
labels = ['A', 'B', 'C', 'D']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.savefig('pie_chart_2.png')
plt.close()

# 5. Histogram
data = np.random.randn(1000)
plt.hist(data, bins=20, color='purple', alpha=0.7)
plt.savefig('histogram_2.png')
plt.close()

# 6. Heatmap
data_matrix = np.random.rand(6, 6)
plt.imshow(data_matrix, cmap='coolwarm', interpolation='nearest')
plt.colorbar(label='Intensity')
plt.savefig('heatmap_2.png')
plt.close()

# 7. Simplified 3D Surface Plot
x = np.linspace(-2, 2, 10)
y = np.linspace(-2, 2, 10)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + y**2)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
plt.savefig('surface_plot_2.png')
plt.close(fig)

# 8. Box Plot
data = [np.random.rand(50), np.random.rand(50) + 1, np.random.rand(50) + 2]
plt.boxplot(data, vert=True, patch_artist=True)
plt.savefig('box_plot_2.png')
plt.close()
```
