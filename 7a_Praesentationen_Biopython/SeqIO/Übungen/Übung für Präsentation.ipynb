```python
from Bio import SeqIO

#Overview of the use of SeqIO 
seq_record = SeqIO.read("organism.fasta","fasta")
#Ouput of the SeqRecord object generated using a fasta file
print(seq_record)


seq_record.id = "Test"
seq_record.name = "Test2"
seq_record.description = "Test3"
seq_record.features = "dies sind Features"
seq_record.seq = "Test5"


print(f"\nSeqRecord after manipulation:\n{seq_record}")
```

```python
#generation of a Iterator using a Genedb file
record_iter = SeqIO.parse("sequence.gb","gb")

#type of the Iterator
print(type(record_iter),end="\n\n")

#next element
seq_rec2 = next(record_iter)
print(seq_rec2,end="\n\n")

#annotation dictionary of a Genedb file 
print(seq_rec2.annotations)
```

```python
#Generate a Dict without the use of SeqIO key is set to the id, value is set to the sequence 
input = "seq2.fasta"

all_seq = {}

with open(input) as file:
    for line in file:
        if line.startswith(">"):
            header = line.replace("\n","").replace(">","")
            header_split = header.split(" ", 1)
            all_seq[header_split[0]] = ""
        else:
            all_seq[header_split[0]] += line.replace("\n","")
        

print(all_seq)
```

```python
#easier to do parse the file using SeqIO
for seq in SeqIO.parse("seq2.fasta", "fasta"):
    print(seq)
    print()
```

```python
#Converting of a genedb file to a fasta file
#Output is the number on sequences in the converted file 
SeqIO.convert("sequence2.gb","gb","seq2.fasta","fasta")
```

```python
#Using .parse() to generate seq objects and gather the amount of sequences with a specific length as well as the total number of sequences in the file
count = 0
total = 0

for sequenz in SeqIO.parse("contigs.fasta", "fasta"):
    if len(sequenz.seq) > 200:
        count += 1
    total += 1

print(f'The number of Sequenzes with a lenght of > 200 is {count} out of {total} total sequenzes')
```

```python
#using .parse() as well as the gzip module to open and parse comperessed files without needing to decompress them
#to optimize memory the use of iterator objects is recomended to reduce the memory use 
import gzip
with gzip.open("organism1.fasta.gz", "rt") as handle:
    for seq in SeqIO.parse(handle, "fasta"):
        print(seq)
```
